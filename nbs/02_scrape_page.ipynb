# Scrape Pages

> Given a url, get rendered html content.


Many, many web sites do not render html directly. Instead, the load a shell and then request most of the data with javascript. The main way to get page data in this situation is a headless browser. This module will use [ScrapingAnt](https://scrapingant.com), a web scraping service that has a generous free tier, with a fallback to [ScrapingBee](https://www.scrapingbee.com/) which may be more reliable sometimes but is more expensive.

```python
#| hide

PAGE_URL = "https://www.ttnews.com/articles/continental-earnings-q4-2023"
```

```python
#| export

import requests
from rich.pretty import pprint
import os
```

```python
#| export

def _scraping_bee_request(url: str, timeout=20):
    response = requests.get(
        url="https://app.scrapingbee.com/api/v1/",
        params={
            "api_key": os.environ['SCRAPING_BEE_API_KEY'],
            "url": url,
        },
        timeout=timeout,
    )
    response.raise_for_status()
    return response.text
```

```python
#| export

def _scraping_ant_request(url, timeout: int = 40):
    response = requests.get(
        url="https://api.scrapingant.com/v2/general",
        params={
            "url": url,
            "x-api-key": os.environ['SCRAPING_ANT_API_KEY'],
        },
        timeout=timeout,
    )
    response.raise_for_status()
    return response.text
```
